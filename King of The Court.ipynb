```python
import numpy as np

def store_the_list(num_of_elements, ordered_list):
# Store the list
  for i in range(num_of_elements):
    element = input(f"Enter element {i+1}: ")
    ordered_list.append(element)
  return ordered_list

# Convert the list into a random list
import numpy as np

def get_random_list(ordered_list, num_of_elements):
    random_list = []
    pick_times = [0] * len(ordered_list)

    while len(random_list) < num_of_elements:
        num = np.random.randint(0, num_of_elements)
        if pick_times[num] == 1:
            continue
        pick_times[num] = 1
        random_list.append(ordered_list[num])

    return random_list


# Compare
def compare(random_list):
    while len(random_list) > 1:
        x = random_list[0]
        y = random_list[1]
        while True:
            ch = input(f"1) {x} VS 2) {y}: ")
            if ch == '1':
                del random_list[1]
                break
            elif ch == '2':
                del random_list[0]
                break
            else:
                print("Enter only 1 or 2!")

    print(f"Final winner: {random_list[0]}")

# Get number of elements
def main():
  num_of_elements = int(input("Enter the number of elements you want to compare: "))

  # Create an array to store the elements in an order
  ordered_list = []
  store_the_list(num_of_elements, ordered_list)
  random_list = get_random_list(ordered_list, num_of_elements)
  compare(random_list)


main()
```
